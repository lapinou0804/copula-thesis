```python
import yfinance as yf
import pandas as pd
```

```python
start_date = "2008-03-01"
end_date   = "2026-01-01"
```

```python

tickers = [
    "SPY", "QQQ", "IWM", "VGK", "EWJ",
    "EEM", "FXI", "EWZ", "EPI", "EFA", 
    "IEF","TLT","LQD","JNK","EMB","PCY",
    "GC=F","CL=F","NG=F","SI=F","DBA",
    "EURUSD=X","JPYUSD=X","GBPUSD=X","AUDUSD=X","CNYUSD=X",
    "VNQ","GLD","VXF","IYR"
    
]
daily = yf.download(
    tickers,
    start=start_date,
    end=end_date,
    interval="1d",
    auto_adjust=False,
    group_by="ticker",
    threads=True
)
```

```python
adj_close = pd.concat(
    [daily[ticker]["Adj Close"] for ticker in tickers],
    axis=1
)

adj_close.columns = tickers
adj_close = adj_close.dropna(how="all")
```

```python
tickers = [
    "SPY", "QQQ", "IWM", "EFA", "EEM",
    "TLT", "IEF", "LQD", "HYG", "SHY",
    "GLD", "SLV", "USO", "DBC", "UNG",
    "VNQ", "XLK", "XLF", "XLE", "XLY",
    "XLP", "XLI", "XLV", "XLB", "XLU",
    "EPI", "FXI", "EWJ", "EWG", "EWU"
]
```

```python
weekly_prices = adj_close.resample("W-FRI").last()
weekly_prices = weekly_prices.loc["2008-03-01":"2025-12-31"]
```
